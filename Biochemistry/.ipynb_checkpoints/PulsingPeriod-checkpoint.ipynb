```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

%matplotlib notebook

data = pd.read_csv('output.csv',delimiter=',')
```

```python
data.head()
```

```python
data.tail()
```

```python
t = data.t
Rm = data.Rm
Am = data.Am
Bm = data.Bm
```

```python
plt.figure(1)
plt.plot(t,Rm)
plt.ylim([0,10])
plt.show()
```

```python
Rm_slope = []
for i in range(0,len(t)-1):
    slope = (Rm[i+1] - Rm[i])/(t[i+1]-t[i])
    Rm_slope.append(slope)
Rm_slope.append(Rm_slope[-1])
```

```python
plt.figure(2)
plt.plot(t[0:len(Rm_slope)],Rm_slope)
plt.xlim([0,500])
plt.ylim([-0.005,0.005])
plt.show()
```

```python
# To Do:  Only keep maximums that are greater than a threshold concentration
max = []
t_max = []
for i in range(0,len(Rm_slope)-1):
    if Rm_slope[i]>0 and Rm_slope[i+1]<0 and Rm[i]>0.1:
        max.append(Rm[i+1])
        t_max.append(t[i+1])

plt.figure(10)
plt.plot(t,Rm)
plt.plot(t_max,max, "o", label="max")
plt.ylim([0,4])
plt.legend()
plt.show()
```

```python
plt.figure(11)
plt.hist(np.diff(t_max))
```
